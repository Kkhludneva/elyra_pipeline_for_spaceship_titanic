# Проанализируем важность признаков

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
catboost = pd.read_csv('model_results/catboost_importances.csv')
logist_reg = pd.read_csv('model_results/logistic_reg_importances.csv')
rand_forest = pd.read_csv('model_results/random_forest_importances.csv')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
imp = [logist_reg['Importance'],rand_forest['Importance'],catboost['Importance']]
imp[0] = sc.fit_transform(imp[0].values.reshape(-1,1))
imp[1] = sc.fit_transform(imp[1].values.reshape(-1,1))
imp[2] = sc.fit_transform(imp[2].values.reshape(-1,1))
plt.plot(logist_reg['Feature'], abs(imp[0]),label='LogReg')
plt.plot(rand_forest['Feature'], abs(imp[1]),label='RandomForest')
plt.plot(catboost['Feature'], abs(imp[2]),label='CatBoost')
plt.title('Feature importances')
plt.legend(loc='best')
plt.xticks(rotation='vertical')
plt.savefig('feature_importances.png')
plt.show()
```
